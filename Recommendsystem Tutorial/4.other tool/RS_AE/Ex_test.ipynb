## Trust_walker(pass)

```python
import sys

sys.path.append(r'RSA/')
sys.path.append(r'RSA')
from RSA import run_func
import numpy as np
```

```python
def init_model(rs):
    rs.read_trust(rs)
    rs.p = 1.0
    rs.k = 0  # 步数

    def get_followee_user(rs, user):
        p = 0
        followees = list(rs.tg.get_followees(user))
        num_foll = len(followees)
        if num_foll == 0:
            return None, 0
        # pick one randomly
        ind = np.random.randint(num_foll)
        p = 1.0 / num_foll  # 选取用户v的概率
        return followees[ind], p

    def get_stop_prob(rs, user, item, k):
        p = 1.0
        sum_sim = 0.0
        max_sim = 0
        max_prob = 0.0
        max_prob_item = 0
        if k == 0:  # if k==0，the stop probability=0
            rs.p = 1.0
            return 0, 0, 0

        param = rs.get_sigmoid_2(k)  # 1/(1+e^(-0.5z))

        u_items = rs.user_rated_items(user)  # item_id list
        print(u_items)
        if len(u_items) == 0:
            return 0, 0, 0
        for u_item in u_items:
            sim = rs.get_sim(rs, item, u_item)
            sum_sim += sim
            prob = sim * param
            if prob > max_prob:
                max_sim = sim
                max_prob = prob
                max_prob_item = u_item
        if sum_sim == 0:
            return 0, 0, 0
        return max_prob, max_prob_item, max_sim / sum_sim  # 返回停止概率，最相似item，选择item j的概率

    def get_sim(rs, item1, item2):
        return rs.get_cosine_improved_sp(rs.get_col(item1), rs.get_col(item2))

    def single_random_walk(rs, user, item, k):
        print(user, item, k)
        print('%s%d' % ('k=', k))
        if rs.containsUserItem(
                user, item
        ):  # judge whether user u rate on item i, if so, return the rating.
            return rs.p, rs.trainSet_u[user][item]
        else:
            rand_num = np.random.rand(1)  # get random number
            # compute the stop probability
            stop_prob, max_item, p_j = rs.get_stop_prob(rs, user, item,
                                                        k)  # p_j 选择item j的概率
            print('stop probability:' + str(stop_prob))
            # print('%s%d'%('stop probbability:',stop_prob))
            # the probability of stopping walk
            print(rand_num, stop_prob)
            if rand_num < stop_prob or k >= 6:  # no more than six steps
                # get the most similar item j, and return r(u,j)
                rating = rs.trainSet_u[user][max_item]
                rs.p = rs.p * stop_prob * p_j
                return (rs.p, rating)
            # compute the probability of next random walk
            else:
                # get next user for random walk
                next_user, tu_prob = rs.get_followee_user(
                    rs, user)  # if user don't have friends in trust network
                print('next step user is:' + str(next_user))
                if next_user == None:  # if no next user
                    _, max_item, p_j = rs.get_stop_prob(rs, user, item,
                                                        -1)  # no sense if k=-1
                    if max_item == 0:  # if no next user and no similar users
                        return rs.p, 0
                    rating = rs.trainSet_u[user][max_item]
                    rs.p = rs.p * p_j
                    return (rs.p, rating)

                rs.p = rs.p * (1 - stop_prob) * tu_prob
                k += 1
                return rs.single_random_walk(rs,
                                             user=next_user,
                                             item=item,
                                             k=k)  # 特么忘了return 当然返回None啦

    rs.get_followee_user = get_followee_user
    rs.get_stop_prob = get_stop_prob
    rs.get_sim = get_sim
    rs.single_random_walk = single_random_walk
```

```python
def predict(rs, u, i):
    user = u
    item = i
    k = rs.k
    prob,pred = rs.single_random_walk(rs, user, item, k)
    return pred
```

```python
run_func(init_model=init_model,predict=predict)
```

## Trust_SVD(pass)

```python
import sys
sys.path.append(r'Z:\RSA')
sys.path.append(r'Z:')
from RSA import run_func
import numpy as np
import math
```

```python
def init_model(rs):
    rs.lambdaB = 1.2
    rs.lambdaY = 1.2
    rs.lambdaW = 1.2
    rs.lambdaT = 0.9
    rs.config.maxIter = 2  # 测试用
    rs.read_trust(rs)
    rs.Bu = np.random.rand(rs.get_train_size()[0]) / (rs.config.factor**0.5
                                                      )  # bias value of user
    rs.Bi = np.random.rand(rs.get_train_size()[1]) / (rs.config.factor**0.5
                                                      )  # bias value of item
    rs.Y = np.random.rand(rs.get_train_size()[1], rs.config.factor) / (
        rs.config.factor**0.5)  # implicit preference
    rs.W = np.random.rand(rs.get_train_size()[0], rs.config.factor) / (
        rs.config.factor**0.5)  # implicit preference

    def get_sum_y(rs, u):
        u_items = rs.user_rated_items(u)
        nu = len(u_items)
        sum_y = np.zeros(rs.config.factor)
        for j in u_items:
            sum_y += rs.Y[rs.item[j]]
        sum_y /= (np.sqrt(nu))
        return nu, sum_y

    def get_sum_w(rs, u):
        followees = rs.tg.get_followees(u)
        nu = 1
        sum_w = np.zeros(rs.config.factor)
        for v in followees.keys():
            if rs.containsUser(v):
                nu += 1
                sum_w += rs.W[rs.user[v]]
        sum_w /= np.sqrt(nu)
        return nu, sum_w

    rs.get_sum_y = get_sum_y
    rs.get_sum_w = get_sum_w
```

```python
def train_model(rs, k):
    iteration = 0
    while iteration < rs.config.maxIter:
        rs.loss = 0
        for index, line in enumerate(rs.trainSet(k)):
            user, item, rating = line
            u = rs.user[user]
            i = rs.item[item]
            error = rating - rs.predict_rs(rs, u=user, i=item)
            rs.loss += error**2

            p, q = rs.P[u], rs.Q[i]
            nu, sum_y = rs.get_sum_y(rs,user)
            nv, sum_w = rs.get_sum_w(rs,user)

            frac = lambda x: 1.0 / math.sqrt(x)

            # update latent vectors
            rs.Bu[u] += rs.config.lr * (
                error - rs.lambdaB * frac(nu) * rs.Bu[u])
            rs.Bi[i] += rs.config.lr * (
                error - rs.lambdaB * frac(nv) * rs.Bi[i])

            rs.Q[i] += rs.config.lr * (error * (p + sum_y + sum_w) -
                                           rs.config.lambdaQ * frac(nu) * q)

            followees = rs.tg.get_followees(user)
            ws = np.zeros(rs.config.factor)
            for followee in followees:
                if rs.containsUser(user) and rs.containsUser(
                        followee):
                    nw = len(rs.tg.get_followers(followee))
                    vid = rs.user[followee]
                    w = rs.W[vid]
                    weight = 1  # followees[followee]
                    err = w.dot(p) - weight
                    rs.loss += err**2
                    ws += err * w
                    rs.W[vid] += rs.config.lr * (
                        err * frac(nv) * q - rs.lambdaT * err * p -
                        rs.lambdaW * frac(nw) * w)  # 更新w
            rs.P[u] += rs.config.lr * (
                error * q - rs.lambdaT * ws -
                (rs.config.lambdaP * frac(nu) +
                 rs.lambdaT * frac(nv)) * p)

            u_items = rs.user_rated_items(u)
            for j in u_items:
                idj = rs.item[j]
                rs.Y[idj] += rs.config.lr * (
                    error * frac(nu) * q -
                    rs.lambdaY * frac(nv) * rs.Y[idj])

        rs.loss += rs.config.lambdaP * (rs.P * rs.P).sum() + rs.config.lambdaQ * (rs.Q * rs.Q).sum() \
                     + rs.lambdaB * (
                             (rs.Bu * rs.Bu).sum() + (rs.Bi * rs.Bi).sum()) + rs.lambdaY * (
                             rs.Y * rs.Y).sum() + rs.lambdaW * (rs.W * rs.W).sum()
        iteration += 1
        if rs.isConverged_rs(iteration, rs, k):
            break
```

```python
def predict(rs, u, i):
    if rs.containsUser(u) and rs.containsItem(i):
        _, sum_y = rs.get_sum_y(rs,u)
        _, sum_w = rs.get_sum_w(rs,u)
        u = rs.user[u]
        i = rs.item[i]
        return rs.Q[i].dot(rs.P[u] + sum_y + sum_w
                             ) + rs.globalMean + rs.Bi[i] + rs.Bu[u]
    elif rs.containsUser(u) and not rs.containsItem(i):
        return rs.userMeans[u]
    elif not rs.containsUser(u) and rs.containsItem(i):
        return rs.itemMeans[i]
    else:
        return rs.globalMeanget_sum_y(rs,u)
```

```python
run_func(init_model=init_model,train_model=train_model,predict=predict)
```

## Social_cune(pass)

```python
import sys
sys.path.append(r'Z:\RSA')
sys.path.append(r'Z:')
from RSA import run_func
import numpy as np
import gensim.models.word2vec as w2v
from collections import defaultdict
from random import choice
```

```python
def init_model(rs):
    rs.read_trust(rs)
    rs.alpha = 0.1
    rs.walkCount = 30
    rs.walkLength = 20
    rs.walkDim = 20
    rs.winSize = 5
    rs.topK = 50
    rs.config.maxIter = 2  # 测试用
    rs.user_sim = rs.get_SimMatrix()

    def generate_cu_net(rs):
        print('Building collaborative user network...')
        itemNet = {}
        for item in rs.trainSet_i:
            if len(rs.trainSet_i[item]) > 1:
                itemNet[item] = rs.trainSet_i[item]

        filteredRatings = defaultdict(list)
        for item in itemNet:
            for user in itemNet[item]:
                if itemNet[item][user] > 0:
                    filteredRatings[user].append(item)

        rs.CUNet = defaultdict(list)

        for user1 in filteredRatings:
            s1 = set(filteredRatings[user1])
            for user2 in filteredRatings:
                if user1 != user2:
                    s2 = set(filteredRatings[user2])
                    weight = len(s1.intersection(s2))
                    if weight > 0:
                        rs.CUNet[user1] += [user2]  # * weight

        # cpprint(rs.CUNet)
        pass
    
    def deep_walk(rs):
        print('Generating random deep walks...')
        rs.walks = []
        rs.visited = defaultdict(dict)
        for user in rs.CUNet:
            for t in range(rs.walkCount):
                path = [str(user)]
                lastNode = user
                for i in range(1, rs.walkLength):
                    nextNode = choice(rs.CUNet[lastNode])
                    count = 0
                    while (nextNode in rs.visited[lastNode]):
                        nextNode = choice(rs.CUNet[lastNode])
                        # break infinite loop
                        count += 1
                        if count == rs.walkLength:  # 10
                            break
                    path.append(str(nextNode))
                    rs.visited[user][nextNode] = 1
                    lastNode = nextNode
                rs.walks.append(path)
        # shuffle(rs.walks)
        # cpprint(rs.walks)
        print('Generating user embedding...')
        rs.model = w2v.Word2Vec(rs.walks, size=rs.walkDim, window=5, min_count=0, iter=3)
        print('User embedding generated.')
        pass
    
    def compute_social_sim(rs):
        print('Constructing similarity matrix...')
        # rs.W = np.zeros((rs.get_train_size()[0], rs.walkDim))
        rs.topKSim = defaultdict(dict)
        i = 0
        for user1 in rs.CUNet:
            sims = {}
            for user2 in rs.CUNet:
                if user1 != user2:
                    wu1 = rs.model[str(user1)]
                    wu2 = rs.model[str(user2)]
                    sims[user2] = rs.get_cosine(wu1, wu2)  # 若为空咋整
            rs.topKSim[user1] = sorted(sims.items(), key=lambda d: d[1], reverse=True)[:rs.topK]
            i += 1
            if i % 200 == 0:
                print('progress:', i, '/', len(rs.CUNet))
        # print(rs.topKSim)
        # 构建被关注列表
        print('Constructing desimilarity matrix...')
        rs.topKSimBy = defaultdict(dict)
        for user in rs.topKSim:
            users = rs.topKSim[user]
            for user2 in users:
                rs.topKSimBy[user2[0]][user] = user2[1]
        print('Similarity matrix finished.')
    
    rs.generate_cu_net = generate_cu_net
    rs.deep_walk = deep_walk
    rs.compute_social_sim = compute_social_sim
```

```python
def train_model(rs, k):
    rs.generate_cu_net(rs)
    rs.deep_walk(rs)
    rs.compute_social_sim(rs)
    iteration = 0
    while iteration < rs.config.maxIter:
        rs.loss = 0
        for index, line in enumerate(rs.trainSet(k)):
            user, item, rating = line
            u = rs.user[user]
            i = rs.item[item]
            error = rating - rs.predict_rs(rs, u=user, i=item)
            rs.loss += 0.5 * error**2
            p, q = rs.P[u], rs.Q[i]

            social_term_p, social_term_loss = np.zeros(
                (rs.config.factor)), 0.0
            followees = rs.topKSim[
                user]  # rs.tg.get_followees(user) #rs.topKSim[user]
            # print(followees)
            for followee in followees:
                if rs.containsUser(followee[0]):
                    # s = rs.user_sim[user][followee]
                    uf = rs.P[rs.user[followee[0]]]
                    social_term_p += followee[1] * (p - uf)
                    social_term_loss += followee[1] * ((p - uf).dot(p - uf))

            social_term_m = np.zeros((rs.config.factor))
            followers = rs.topKSimBy[user]
            followers = sorted(followers.items(),
                               key=lambda d: d[1],
                               reverse=True)[:rs.topK]
            for follower in followers:
                if rs.containsUser(follower[0]):
                    ug = rs.P[rs.user[follower[0]]]
                    social_term_m += follower[1] * (p - ug)

                    # update latent vectors
            rs.P[u] += rs.config.lr * (error * q - rs.alpha *
                                           (social_term_p + social_term_m) -
                                           rs.config.lambdaP * p)
            rs.Q[i] += rs.config.lr * (error * p - rs.config.lambdaQ * q)

            rs.loss += 0.5 * rs.alpha * social_term_loss

        rs.loss += 0.5 * rs.config.lambdaP * (rs.P * rs.P).sum(
        ) + 0.5 * rs.config.lambdaQ * (rs.Q * rs.Q).sum()

        iteration += 1
        if rs.isConverged_rs(iteration,rs,k):
            break
```

```python
run_func(init_model=init_model,train_model=train_model)
```
