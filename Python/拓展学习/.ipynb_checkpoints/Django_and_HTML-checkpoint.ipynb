##### 路由
- 按照具体的请求url,导入到相应的业务处理模块的一个功能模块
- django的信息控制中枢
- 本质上是接受的URL和相应处理函数的一个映射
- 在接受URL请求的匹配上使用了RE
- URL的具体格式在urls.py中显示
- 子路由用include
- 可以输入参数
- 反向解析reverse()

```
url(r'index/((page-)?(?P<oage_number>\d+)/)?$',view.myindex)
```

##### views视图
- 接收web请求并返回web响应的事物处理函数
- 包括json，string，html等
- 重定向HttpResponseRedirect
- Request的属性
    - path
    - method
    - encoding
    - GRT
    - POST
    - FILES
    - COOKIES
    - session
- is_ajax():如果是XMLHttpRequest发起的则是True
- GET,POST都是QueryDict对象，类似dict但一个键可以带多个值
    - get()获取最后一个值
    - getlist()获取一个列表
- GET属性
    - 一般在？后面
    - 用key1=value1表示
    - 参数间用&相连
- POST属性
    - 必须有name属性，CheckBox存在一键多值的问题
- 设置报错404
    - default.page_not_found(request,template_name="404.html"
    
##### Models模型
- ORM:ObjectRelationMap
- models.py中定义class
- 所有class必须是models.Model的子类
- 字段类型必须用models.XXX
- 负责与数据库交互

- 在setting里切换数据库
    - mysql
    
    ```
    DATABASES = [
    "default" = {
        'ENGINE': 'django.db.backends.mysql',
        'NAME':'achlier',
        'USER':'root',
        'PASSWORD':'123456',
        'HOST':'127.0.0.1',
        'POST':'3306',
        }
    ]
    ```
    
    - 同时在主项目的__init__文件中
    
    ```
    import pymysql
    pymysql.install_as_MySQLdb()
    ```
 
- 字段常用参数
    - max_length:最大长度
    - blank:是否允许字段为空
    - null
    - default
    - unique
    - verbose_name
- 常见filter方法
    - gt:大于
    - gte:大于等于
    - lt:小于
    - lte:小于等于
    - range:范围
    - year:年份
    - isnull:是否为空
    - exact:精确等于
    - iexact:不区分大小写 #Town.objects.filter(name__iexact= "storytown" )
    - contains:包含
    - startwith:以..开头
    - endwith:以..结尾
- one_to_one: models.OneToOneField(类名,on_delete=models.CASCADE)
- one_to_many: models.ForeignKey(类名,on_delete=models.CASCADE)
- many_to_many: models.ManyToManyField(类名,on_delete=models.CASCADE)
- 数据库的迁移
    > python .\manage.py makemigrations
    
    > python manage.py migrate
- 启动命令行: 
    > python .\manage.py shell
    
    > from 应用.models import 类名 #from chara.models import Town,Governor,Power,Individual
