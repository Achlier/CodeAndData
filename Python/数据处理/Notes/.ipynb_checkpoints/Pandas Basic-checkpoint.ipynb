**Slicing by labels:** df.loc[row_label,column_label]

* `df.loc`'s first argument is row label. Its second argument is column label, which is optional. Both of the two arguments could be single label, or a list of labels. If both arguments are list, then return a DataFrame, else return a Series. 

````python
df1.loc['2017-11-02']
````

````python
df1.loc[:,'A']
````

````python
df1.loc['2017-11-02':'2017-11-05','A':'C']
````

**Slicing by numbered position:** df.iloc[row_number(s), column_number(s)]

````python
df1.iloc[0,1]
````

````python
df1.iloc[:,1]
````

````python
df1.iloc[1:3,2:4]
````

````python
df1.iloc[[0,2], [1,3]]
````

**intelligent slicing: ** df.ix[]

* Pandas will check the input and intelligently decide to use either labeled slicing or numbered position slicing.

````python
df1.ix[1, 'A':'C']
````

````python
df1.ix['2017-11-02':'2017-11-05', 1:3] 
````

**Slicing by boolean variable**

* Dataframe can be by boolean list, like numpy array

````python
df1[df1.A>0]
````

````python
df1[(df1.A>0) & (df1.B<0)]
````

**Dataframe slicing result is a reference, not copy**

````python
web_stats = { 'Code' : '000001',
              'Open' : [11.56,11.36,11.49,11.42], 
              'High' : [11.59, 11.58, 11.68, 11.42],
              'Close': [11.40,11.54,11.39,11.28]
            }
df2 = pd.DataFrame(web_stats,columns=['Code', 'Open', 'High', 'Close' ], index=pd.date_range('20171101',periods=4))
df2
````

````python
df2.loc[0:2, :]=0
df2
````

**You can chain the slicing **

````python
df2
````

````python
df2.loc['2017-11-01':'2017-11-03', ['Code', 'Open', 'High']].loc[:, ['Code', 'Open']]
````

## Handling NaN values

````python
df1 = pd.DataFrame(np.random.randn(6,4), index=dates,columns=['A','B','C','D'])
df1
````

````python
df1.loc['2017-11-01', 'A'] = np.NaN
df1.loc['2017-11-04', 'B'] = np.NaN
df1
````

* `dropna` will return a **copy** of droped DataFrame

````python
df1.dropna() # return a copy
````

````python
df1 # original DataFrame is not changed
````

````python
df1.fillna(0)
````

````python
df1
````

* **Fill NaN values with column mean**

````python
df1.mean(axis=0)
````

````python
df5 = df1.apply(lambda x: x.fillna(x.mean()), axis=0)
df5
````

It specifies the axis along which the means are computed. By default axis=0. This is consistent with the numpy.mean usage when axis is specified explicitly (in numpy.mean, axis==None by default, which computes the mean value over the flattened array) , in which axis=0 along the rows (namely, index in pandas), and axis=1 along the columns.
```
+------------+---------+--------+
|            |  A      |  B     |
+------------+---------+---------
|      0     | 0.626386| 1.52325|----axis=1----->
+------------+---------+--------+
             |         |
             | axis=0  |
             ↓         ↓
```

````python
df1.loc['2017-11-01', 'A'] = np.NaN
df1.loc['2017-11-04', 'B'] = np.NaN
df1
````

* **Fill NaN with row mean**

````python
df1.mean(axis=1)
````

````python
df5 = df1.apply(lambda x: x.fillna(x.mean()), axis=1)
df5
````

## Combine Multiple DataFrame

````python
index = pd.date_range('2017-11-04', periods=6)
df3 = pd.DataFrame(data=np.random.randn(6,4), index=index, columns=list('ABCD'))
df3
````

````python
df4 = pd.concat([df1, df3], axis=0) # you can see there has duplicates there.
df4
````

````python
pd.concat([df1, df3], axis=1)
````
