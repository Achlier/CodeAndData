# 1. GroupBy

* Intuitively, you want to split the dataset into groups, one for each stock, and then to compute a summary statistic of the stock over years, such as the mean or the median, and then to see whether this statistic increases over the years.
* Pandas's `groupby` method helps you group records together by certain attributes, API can be found [here](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.groupby.html).

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})
```

```python
df
```

**Groupby w.r.t stock code, obtain the `groupby` object**

```python
grouped = df.groupby('A')
type(grouped)
```

```python
grouped.groups
```

```python
grouped = df.groupby(['A', 'B'])
grouped.groups
```

**With the GroupBy object in hand, iterating through the grouped data is very natural**

```python
for name, gp in grouped:
    print (name)
    print (gp)
    print (type(gp))
```

**You can select a specific group**

```python
grouped = df.groupby('A')
agroup = grouped.get_group('foo')
agroup
```

**Summary statistics of each column**

```python
df.groupby('A').describe()
```

**You can choose to view a specific column's summary statistics**

```python
df.groupby('A').C.describe()
```

#### Get median of each column within each group

```python
df.groupby('A').median() # Try max(), min(), std(), mean()
```

**You may not choose the groupby column as index (compare to above)**

```python
df.groupby('A', as_index=False).median() 
```

#### Count each stock's number of trading days

```python
df.groupby('A').count()
```

```python
df['A'].value_counts()
```

#### Aggregate summary statistics over comumn "C"

```python
df.groupby('A').C.aggregate(['count', 'min', 'max', 'median'])
```

# 2. Split-Apply-Combine strategy

* `groupby` is far more powerful than only calculate some summary statistics of the group.
*  In fact, in many situations you may wish to **split** the data set into groups and **do something** with those groups yourself, then **combine** the processed groups together.
* This is commonly called `split-apply-combine` strategy.
* Generally, it involves 3 steps:
    1. `Splitting` the data into groups based on some criteria
    2. `Applying` a function to each group independently
    3. `Combining` the results into a data structure

**Split**

The split step is the most straightforward. It justs sperates a dataframe into splits according to some column(s).

**Apply**

In the apply step, we might wish to one of the following:
* `Aggregation`: computing a summary statistic (or statistics) about each group. Some examples:
    * Compute group sums or means
    * Compute group sizes / counts
* `Transformation`: perform some group-specific computations and return a like-indexed. Some examples:
    * Standardizing data (zscore) within group
    * Filling NAs within groups with a value derived from each group
* `Filtration`: discard some groups, according to a *group-wise* computation that evaluates *True* or *False*. Some examples:
    * Discarding data that belongs to groups with only a few members
    * Filtering out data based on the group sum or mean

**Combine**

GroupBy will examine the results of the apply step and try to return a sensibly combined result.

### 2. Transformation

* Different to groupby `aggregation` that returns a scalar (a number) of a group, the `transform` method returns an object that is indexed the **same size** as the one being grouped. 

what the transform function is passed is not the entire group as a dataframe, but a single column of a single group.

**Example: Compare groupby `Aggregation` and `Transformation`**

```python
df = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
                   'B' : ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
                   'C' : np.random.randn(8),
                   'D' : np.random.randn(8)})

df
```

* We can see each group is associated with a scalar (mean), e.g. "foo one" group has two records, and the group's "C" column 0.520487 is obtained by the average of the two records (0.398759+0.895289)/2.

```python
df.groupby(['A'],as_index=False).agg(np.mean)
```

* We can see row 0,2,,4,6,7 ("foo") and row 1,3,5 ("bar") have the same value on C and D column, this is because the scalar is expanded to all rows of records of the group. 
* transform function "np.mean" is applied to group("foo")\[C], group("foo")\[D], group("boo")\[C], group("boo")\[D] respectively, rather than work as np.mean(group("foo")) and np.mean(group("boo"))

```python
df.groupby(['A']).transform(np.mean)
```

**Example: Sample Mean Distribution**

If $X_1, X_2, ... , X_n$ are observations of a random sample of size n from a $N(\mu, \sigma^2)$ population, then the sample mean: 

\begin{equation}
\bar{X} = \frac{1}{n} \sum\limits_{i=1}^n X_i 
\end{equation}

is normally distributed with mean $\mu$ and variance $\sigma^2/n$. That is, the probability distribution of the sample mean is: $N(\mu,\sigma^2/n)$

```python
index = pd.date_range('01/01/2010',periods=1100)
ts = pd.Series(np.random.normal(0.5,2,1100), index)
ts_roll = ts.rolling(window=100,min_periods=100).mean().dropna()
```

```python
pd.DataFrame({'ts': [ts.mean(), ts.std(), len(ts)], 'ts_roll': [ts_roll.mean(), ts_roll.std(), \
                                    len(ts_roll)]}, index=['mean', 'std', 'count'])
```

```python
ts.head()
```

```python
ts.tail()
```

```python
key = lambda x: x.year
ts_roll_transformed = ts_roll.groupby(key).transform(lambda x: (x-x.mean())/x.std())
```

```python
ts_roll_grouped = ts_roll.groupby(key)
ts_roll_grouped.agg(['mean','std', 'count'])
```

```python
transformed_grouped = ts_roll_transformed.groupby(key)
transformed_grouped.agg(['mean','std', 'count'])
```

```python
%matplotlib inline
compare = pd.DataFrame({'Original': ts_roll, 'Transformed': ts_roll_transformed})
compare.plot()
```

**Example: replace missing data with the group mean**

```python
df = pd.DataFrame(np.random.rand(30).reshape(10,3), columns=['A', 'B', 'C'])
df.loc[4, 'A']=np.NAN
df.loc[6, 'B']=np.NAN
df.loc[1, 'C']=np.NAN
df
```

```python
countries = np.array(['US', 'UK', 'GR', 'JP'])
key = countries[np.random.randint(0, 4, 10)]
df_filled = df.groupby(key).transform(lambda x: x.fillna(x.mean()))
```

```python
df.groupby(key).agg(['mean', 'count'])
```

```python
df_filled.groupby(key).agg(['mean', 'count'])
```

## 3. Filter

* The filter method returns a **subset** of the original object which satisfy some certain criteria. A boolean variable must be mapped to each group. Group associated with False will be filtered out.

**Example: select stocks whose average trading volume is large than 10e5**

```python
stocks_avg_vol_10e5 = ten_stocks_daily_bar.groupby('code').filter(lambda x: np.mean(x['volume'])>3*10e4)
```

```python
gpd = stocks_avg_vol_10e5.groupby('code')
```

```python
gpd.count()
```

## 4. Flexible `apply`

* Some operations on the grouped data might not fit into either the aggregate or transform categories. 
* Different to `transform`, the `apply` function will operate on each group's entire **dataframe**, rather than each **column** of the dataframe.

```python
# get each stock's summary statistics
ten_stocks_daily_bar.groupby('code').apply(lambda x: x.describe())
```

**Example: get first 3 records of each stock**

```python
first_three = ten_stocks_daily_bar.groupby('code',as_index=False).apply(lambda x: x[['code','open','close']][0:3]).reset_index(drop=True)
```

```python
first_three.head()
```

**Example: Apply function to each group dataframe**

```python
grouped = ten_stocks_daily_bar.groupby('code')['volume']
def func(group):
    return pd.DataFrame({'original': group, 'demean':group-np.mean(group)})


gp = grouped.apply(func)
print (gp.head())
print (gp.shape)
```

```python
grouped = ten_stocks_daily_bar.groupby('code')['volume']
gp = grouped.apply(lambda x: pd.DataFrame({'original': x, 'demean':x-np.mean(x)}))
print (gp.head())
print (gp.shape)
```

```python
grouped = ten_stocks_daily_bar.groupby('code')['volume']
df = pd.DataFrame()
for name, group in grouped:
    df = df.append(pd.DataFrame({'original': group, 'demean':group-np.mean(group)}))
print (df.head())
print (df.shape)
```

* The three examples functions exactly the same.
* We can find that groupby `apply` function takes a groupby dataframe as input, perform some operation, then return the processed dataframe. All these returned dataframes are joined together.
* `apply` can act as a reducer, transformer, or filter function, depending on exactly what is passed to it. So depending on the path taken, and exactly what you are grouping.
