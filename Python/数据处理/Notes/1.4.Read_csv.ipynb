### Change date format

* In case the `date` column is not in appropriate format, we can parse it through `date_parser` argument.

```python
df = pd.read_csv('../data/ten_stocks_daily_bar_sub.csv')
df.head()
```

* Suppose we do not like the date format and want to change it to standard format.

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y')

df = pd.read_csv('../data/ten_stocks_daily_bar_sub.csv', parse_dates=['date'], date_parser=dateparse)
df.head()
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y')
print(dateparse('08/01/2017'))
```

### Boolean Slicing

* Find the number of days stock code 1's trading volume is larger than 10^6

```python
code1 = ten_stocks_daily_bar['code']==1
vol = ten_stocks_daily_bar['volume']>1e6
df = ten_stocks_daily_bar[code1&vol]
len(df)
```

### value_counts

```python
stocks_info = pd.read_csv('../data/stocks_info.csv')
```

```python
stocks_info.head()
```

* Top 10 provinces ranked w.r.t the number of listed companies.

```python
top10_provinces = stocks_info['area'].value_counts()[:10]
top10_provinces
```

```python
top10_provinces.plot(kind = 'bar')
```

* Top 10 industries.

```python
top10_industry = stocks_info['industry'].value_counts()[:10]
```

```python
top10_industry.plot(kind='bar')
```
