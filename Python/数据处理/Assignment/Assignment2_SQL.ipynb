**13 Find the number of the companies with "科技" in its name and located in Guangdong province from stocks_info and calculate its average totalAssets.**

13.1 Use SQL.

```python
%sql SELECT AVG(totalAssets) FROM df_stocks_info where area='广东' and name LIKE '%科技%';
```

13.2 Use Pandas Dataframe.

```python
mask=np.array([i[:2]=='科技' for i in df_stocks_info.groupby('area').get_group('广东').name])
```

```python
mask2=np.array([i[-2:]=='科技' for i in df_stocks_info.groupby('area').get_group('广东').name])
```
